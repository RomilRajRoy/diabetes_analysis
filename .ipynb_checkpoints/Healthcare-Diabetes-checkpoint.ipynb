```python
import pandas as pd

# Load the dataset to understand its structure
file_path = 'D:/Healthcare-Diabetes.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure and content
data.head()
```

```python
## Data pre-processing

import numpy as np

# Replacing 0s with NaN for the specified columns
columns_with_zero_as_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for column in columns_with_zero_as_missing:
    data[column] = data[column].replace(0, np.nan)

# Verifing the replacement
print(data.isnull().sum())
```

```python
# Filling NaNs with the median of respective columns
for column in columns_with_zero_as_missing:
    data[column] = data[column].fillna(data[column].median())

# Verifing the cleaning
print(data.isnull().sum())
```

```python
# Handling outliers using the IQR method
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else lower_bound if x < lower_bound else x)

# Appling the outlier handling to the relevant columns
for column in columns_with_zero_as_missing:
    handle_outliers(data, column)

# Verify the changes
data.describe()
```

```python
#Apropri ALgorithm
import pandas as pd
from itertools import combinations

# Discretize continuous data into categorical data
discretized_data = data.copy()
discretized_data['Glucose'] = pd.cut(data['Glucose'], bins=3, labels=['Low', 'Medium', 'High'])
discretized_data['BloodPressure'] = pd.cut(data['BloodPressure'], bins=3, labels=['Low', 'Medium', 'High'])
discretized_data['SkinThickness'] = pd.cut(data['SkinThickness'], bins=3, labels=['Low', 'Medium', 'High'])
discretized_data['Insulin'] = pd.cut(data['Insulin'], bins=3, labels=['Low', 'Medium', 'High'])
discretized_data['BMI'] = pd.cut(data['BMI'], bins=3, labels=['Low', 'Medium', 'High'])
discretized_data['Age'] = pd.cut(data['Age'], bins=3, labels=['Young', 'Middle-aged', 'Old'])

# Convert categorical columns to one-hot encoding
one_hot_data_with_outcome = pd.get_dummies(discretized_data)

# Function to calculate support for individual items including 'Outcome'
def calculate_support(data, item):
    if isinstance(item, list):
        return data[item].all(axis=1).mean()
    return data[item].mean()

# Function to generate frequent itemsets ensuring 'Outcome' is considered
def generate_frequent_itemsets_with_outcome(data, min_support):
    itemsets = []
    for item in data.columns:
        support = calculate_support(data, item)
        if support >= min_support:
            itemsets.append((item, support))
    return itemsets

# Function to generate association rules ensuring 'Outcome' is included
def generate_association_rules_with_outcome(data, itemsets, min_confidence):
    rules = []
    for itemset in itemsets:
        item, support = itemset
        if item != 'Outcome':
            combined_support = data[[item, 'Outcome']].all(axis=1).mean()
            confidence = combined_support / support
            if confidence >= min_confidence:
                rules.append((item, 'Outcome', confidence))
    return rules

# Define adjusted support and confidence thresholds
min_support_tuned = 0.05  
min_confidence_tuned = 0.5 

# Find frequent itemsets with the tuned thresholds
frequent_itemsets_tuned = generate_frequent_itemsets_with_outcome(one_hot_data_with_outcome, min_support_tuned)

# Generate association rules with the tuned thresholds
rules_tuned = generate_association_rules_with_outcome(one_hot_data_with_outcome, frequent_itemsets_tuned, min_confidence_tuned)

# Convert the new rules to a DataFrame for better readability
rules_df_tuned = pd.DataFrame(rules_tuned, columns=['Antecedent', 'Consequent', 'Confidence'])

# Display all new rules
rules_df_tuned

rules_df = pd.DataFrame({
    'Antecedent': ['Glucose_High', 'BMI_High', 'Age_Middle-aged'],
    'Consequent': ['Outcome', 'Outcome', 'Outcome'],
    'Confidence': [0.720293, 0.525381, 0.569966]
})
```

```python
import pandas as pd
from itertools import combinations
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Assuming `rules_df_tuned` contains the tuned rules

# Function to predict the outcome using the rules
def predict_with_rules(data, rules):
    predictions = pd.Series(0, index=data.index)
    for index, rule in rules.iterrows():
        antecedent = rule['Antecedent']
        consequent = rule['Consequent']
        confidence = rule['Confidence']
        if antecedent in data.columns:
            predictions[data[antecedent] == 1] = 1
    return predictions

# Predict outcomes using the rules
y_pred_rules = predict_with_rules(one_hot_data_with_outcome, rules_df_tuned)
y_true = one_hot_data_with_outcome['Outcome']

# Calculate evaluation metrics
accuracy = accuracy_score(y_true, y_pred_rules)
precision = precision_score(y_true, y_pred_rules)
recall = recall_score(y_true, y_pred_rules)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')

# Function to predict probabilities using the rules
def predict_prob_with_rules(data, rules):
    probabilities = pd.Series(0.0, index=data.index)
    for index, rule in rules.iterrows():
        antecedent = rule['Antecedent']
        confidence = rule['Confidence']
        if antecedent in data.columns:
            probabilities[data[antecedent] == 1] = confidence
    return probabilities

# Predict probabilities using the rules
y_prob_rules = predict_prob_with_rules(one_hot_data_with_outcome, rules_df_tuned)

# Calculate ROC curve
fpr, tpr, _ = roc_curve(y_true, y_prob_rules)
roc_auc = roc_auc_score(y_true, y_prob_rules)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#K-fold Cross validation
from sklearn.preprocessing import StandardScaler

# Features and target
X = data.drop(columns=['Id', 'Outcome'])
y = data['Outcome']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# Define the models
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd

# Function to evaluate models using cross-validation
def evaluate_models(models, X, y):
    results = {}
    for name, model in models.items():
        # Using 5-fold cross-validation
        accuracy = cross_val_score(model, X, y, cv=5, scoring='accuracy').mean()
        precision = cross_val_score(model, X, y, cv=5, scoring='precision').mean()
        recall = cross_val_score(model, X, y, cv=5, scoring='recall').mean()
        results[name] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall
        }
    return results

# Evaluate the models
results = evaluate_models(models, X_scaled, y)

# Display the results
results_df = pd.DataFrame(results)
```

```python
def predict_with_rules(data, rules):
    predictions = pd.Series(0, index=data.index)
    for index, rule in rules.iterrows():
        antecedent = rule['Antecedent']
        predictions |= data[antecedent]
    return predictions

def evaluate_rule_based(data, y_true, rules):
    y_pred = predict_with_rules(data, rules)
    accuracy = (y_true == y_pred).mean()
    precision = ((y_pred & y_true).sum() / y_pred.sum()) if y_pred.sum() != 0 else 0
    recall = ((y_pred & y_true).sum() / y_true.sum()) if y_true.sum() != 0 else 0
    return {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall}

# Evaluate the rule-based algorithm
rule_based_results = evaluate_rule_based(one_hot_data_with_outcome, one_hot_data_with_outcome['Outcome'], rules_df_further_adjusted)

# Add the rule-based results to the results DataFrame
results = {
    'Rule-Based': rule_based_results
}

# Display the results
results_df = pd.DataFrame(results).T
print(results_df)
```
